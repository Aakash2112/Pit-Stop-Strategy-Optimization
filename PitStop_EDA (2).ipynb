```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
pit = pd.read_csv('/content/Pit_Stop2.csv')
pit
```

```python
print(pit.info())
print(pit.head())
```

```python
# Check for null values
print("\nNull Values:")
print(pit.isnull().sum())

# Summary statistics for numerical columns
print("\nSummary Statistics:")
print(pit.describe())

# --- Correlation Analysis ---
# Select numeric columns for correlation, excluding certain columns that may not be relevant
num_cols = pit.select_dtypes(include=['int64', 'float64']).columns.drop(
    ['IsWet', 'IsSafetyCarLap', 'IsVSCLap', 'IsNeutralizedLap']
)
correlation = pit[num_cols].corr()

# --- Individual Graphs ---
# 1. Lap Time Distribution (Histogram)
plt.figure(figsize=(10, 6))
sns.histplot(pit["LapTime(s)"], bins=50, kde=True, color='purple', edgecolor='black')
plt.title("Lap Time Distribution", fontsize=14, color='darkblue')
plt.xlabel("Lap Time (s)", fontsize=12)
plt.ylabel("Frequency", fontsize=12)
plt.show()

# 2. Lap Time by Compound (Violin Plot)
plt.figure(figsize=(10, 6))
sns.violinplot(data=pit, x="Compound", y="LapTime(s)", palette="Set2")
plt.title("Lap Time Distribution by Tyre Compound", fontsize=14, color='darkblue')
plt.xticks(rotation=45)
plt.xlabel("Tyre Compound", fontsize=12)
plt.ylabel("Lap Time (s)", fontsize=12)
plt.show()

# 4. Stint Degradation Slope Distribution (Histogram)
plt.figure(figsize=(10, 6))
sns.histplot(pit["StintDegradationSlope"], kde=True, color='green', edgecolor='black')
plt.title("Stint Degradation Slope Distribution", fontsize=14, color='darkblue')
plt.xlabel("Stint Degradation Slope", fontsize=12)
plt.ylabel("Frequency", fontsize=12)
plt.show()

# 5. Lap Time vs Lap Number (Line Plot)
plt.figure(figsize=(10, 6))
sns.lineplot(data=pit, x="LapNumber", y="LapTime(s)", hue="IsWet", palette='coolwarm', linewidth=2)
plt.title("Lap Time vs Lap Number (Wet Highlighted)", fontsize=14, color='darkblue')
plt.xlabel("Lap Number", fontsize=12)
plt.ylabel("Lap Time (s)", fontsize=12)
plt.show()

# 6. Air Temperature vs Lap Time (Scatter Plot)
plt.figure(figsize=(10, 6))
sns.scatterplot(data=pit, x="AirTemp", y="LapTime(s)", hue="Compound", palette='viridis', s=100)
plt.title("Air Temperature vs Lap Time", fontsize=14, color='darkblue')
plt.xlabel("Air Temperature (°C)", fontsize=12)
plt.ylabel("Lap Time (s)", fontsize=12)
plt.legend(title="Tyre Compound")
plt.show()

# 7. Lap in Stint vs Average Lap Time (Line Plot)
plt.figure(figsize=(10, 6))
sns.lineplot(data=pit, x="LapInStint", y="AvgLapTimeInStint", hue="Compound", palette='coolwarm', linewidth=2)
plt.title("Lap in Stint vs Average Lap Time", fontsize=14, color='darkblue')
plt.xlabel("Lap in Stint", fontsize=12)
plt.ylabel("Average Lap Time (s)", fontsize=12)
plt.show()

# 8. Correlation Heatmap (Heatmap)
plt.figure(figsize=(10, 8))
sns.heatmap(correlation, annot=True, cmap="YlGnBu", fmt=".2f", linewidths=1, linecolor='white')
plt.title("Correlation Heatmap", fontsize=14, color='darkblue')
plt.show()

# --- Grouped Analysis for Categorical Data ---
# 9. Lap Time Distribution by Team (Box Plot)
plt.figure(figsize=(12, 6))
sns.boxplot(data=pit, x="Team", y="LapTime(s)", palette='Set1')
plt.title("Lap Time Distribution by Team", fontsize=14, color='darkblue')
plt.xticks(rotation=90)
plt.xlabel("Team", fontsize=12)
plt.ylabel("Lap Time (s)", fontsize=12)
plt.show()

# 10. Average Lap Time by Driver (Bar Plot)
plt.figure(figsize=(12, 6))
sns.barplot(data=pit, x="Driver", y="LapTime(s)", estimator='mean', palette='plasma')
plt.title("Average Lap Time by Driver", fontsize=14, color='darkblue')
plt.xticks(rotation=90)
plt.xlabel("Driver", fontsize=12)
plt.ylabel("Average Lap Time (s)", fontsize=12)
plt.show()

# 11. Lap Time Distribution by Position (Violin Plot)
plt.figure(figsize=(12, 6))
sns.violinplot(data=pit, x="Position", y="LapTime(s)", palette='Set3')
plt.title("Lap Time Distribution by Position", fontsize=14, color='darkblue')
plt.xlabel("Position", fontsize=12)
plt.ylabel("Lap Time (s)", fontsize=12)
plt.show()

# 12. Stacked Bar Chart

# Count values for each column
safety = pit["IsSafetyCarLap"].value_counts()
vsc = pit["IsVSCLap"].value_counts()
neutral = pit["IsNeutralizedLap"].value_counts()

# Combine into a DataFrame
safety_data = pd.DataFrame({
    'Safety Car': safety,
    'Virtual SC': vsc,
    'Neutralized': neutral
}).T  # Transpose for stacking

# Plot
safety_data.plot(kind='bar', stacked=True, color=['#66b3ff', '#ff9999'], figsize=(10, 6), edgecolor='black')
plt.title("Laps by Safety Conditions", fontsize=14, color='darkblue')
plt.xlabel("Condition Type", fontsize=12)
plt.ylabel("Lap Count", fontsize=12)
plt.legend(["False (No)", "True (Yes)"])
plt.xticks(rotation=0)
plt.show()



# 14. Air Temperature vs Humidity (Scatter Plot)
plt.figure(figsize=(10, 6))
sns.scatterplot(data=pit, x="AirTemp", y="Humidity", hue="Compound", palette='coolwarm', s=100, edgecolor='black')
plt.title("Air Temperature vs Humidity", fontsize=14, color='darkblue')
plt.xlabel("Air Temperature (°C)", fontsize=12)
plt.ylabel("Humidity (%)", fontsize=12)
plt.show()

# 15. Track Temperature vs Lap Time (Scatter Plot)
plt.figure(figsize=(10, 6))
sns.scatterplot(data=pit, x="TrackTemp", y="LapTime(s)", hue="IsWet", palette='coolwarm', s=100)
plt.title("Track Temperature vs Lap Time", fontsize=14, color='darkblue')
plt.xlabel("Track Temperature (°C)", fontsize=12)
plt.ylabel("Lap Time (s)", fontsize=12)
plt.show()

# 16. Average laptime by position
# Step 1: Filter out rows with missing LapTime(s) or Position
plot_data = pit[["Position", "LapTime(s)"]].dropna()

# Step 2: Convert 'Position' to int (if not already)
plot_data["Position"] = plot_data["Position"].astype(int)

# Step 3: Create the bar plot
plt.figure(figsize=(12, 6))
sns.barplot(data=plot_data, x="Position", y="LapTime(s)", estimator='mean', palette='inferno')
plt.title("Average Lap Time by Position", fontsize=14, color='darkblue')
plt.xlabel("Position", fontsize=12)
plt.ylabel("Average Lap Time (s)", fontsize=12)
plt.xticks(rotation=45)
plt.show()













```
