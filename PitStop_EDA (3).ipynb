```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
pit = pd.read_csv('/content/Pit_Stop2.csv')
pit
```

```python
print(pit.info())
print(pit.head())
```

```python
# Check for null values
print("\nNull Values:")
print(pit.isnull().sum())

# Summary statistics for numerical columns
print("\nSummary Statistics:")
print(pit.describe())

# --- Correlation Analysis ---
# Select numeric columns for correlation, excluding certain columns that may not be relevant
num_cols = pit.select_dtypes(include=['int64', 'float64']).columns.drop(
    ['IsWet', 'IsSafetyCarLap', 'IsVSCLap', 'IsNeutralizedLap']
)
correlation = pit[num_cols].corr()

# Average Lap Time by Driver (Bar Plot)
plt.figure(figsize=(12, 6))
sns.barplot(data=pit, x="Driver", y="LapTime(s)", estimator='mean', palette='plasma', errorbar=None)
plt.title("Average Lap Time by Driver", fontsize=14, color='darkblue')
plt.xticks(rotation=90)
plt.xlabel("Driver", fontsize=12)
plt.ylabel("Average Lap Time (s)", fontsize=12)
plt.show()


# Air Temperature vs Humidity (Scatter Plot)
plt.figure(figsize=(10, 6))
sns.scatterplot(data=pit, x="AirTemp", y="Humidity", hue="Compound", palette='coolwarm', s=100, edgecolor='black')
plt.title("Air Temperature vs Humidity", fontsize=14, color='darkblue')
plt.xlabel("Air Temperature (°C)", fontsize=12)
plt.ylabel("Humidity (%)", fontsize=12)
plt.show()

```
